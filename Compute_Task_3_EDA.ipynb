```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/haberman.csv')
df
```

```python
df.columns=['patient_age','operation_year','positive_axillary_nodes','survival_status'] #naming columns for thr given dataset
df 
```

```python
df['survival_status'].value_counts() #to check the survival status of various passengers 
```

```python
df.isnull() #no null values present
```

```python
df.isnull().sum() #no null values present
```

```python
sns.heatmap(df.isnull(),  yticklabels = False, cmap = 'BuGn_r') #heat map showing if any column has a null value
```

```python
df['survival_status']=df['survival_status'].map({1:"yes",2:"no"})
```

```python
df.describe() #describes the data in the dataset except 'survival_rate', since its a binary attribute and describe retuns all the numeric calculations
```

```python
df.head() #returns 1st 5 rows of the dataset
```

```python
survival_yes=df[df['survival_status']=="yes"] #creating a new attribute which saves the data of the ones who survived
survival_yes.describe() #complete overview of the data of those who survived
```

```python
survival_no=df[df['survival_status']=="no"] #creating a new attribute which saves the data of the ones who didn't survive
survival_no.describe() #complete overview of the data of those who didn't survive
```

```python
sns.FacetGrid(df, hue = "survival_status").map(sns.distplot, "patient_age").add_legend()
plt.show() #throughout the distribution of age, the plot shows if the person has survived or not
```

```python
sns.FacetGrid(df, hue = "survival_status").map(sns.distplot, "operation_year").add_legend()
plt.show() #over the years, this plot shows the survival of the patient
```

```python
sns.FacetGrid(df, hue = "survival_status").map(sns.distplot, "positive_axillary_nodes").add_legend()
plt.show() #over the distribution of the number of auxillary nodes, this plot shows the survival of the patient
```

```python
plt.figure(figsize = (15, 4))
plt.subplot(1,3,1)
sns.boxplot(x = 'survival_status', y = 'patient_age', data = df) #box-plot showing a relation between the patient's age and survival status
plt.subplot(1,3,2)
sns.boxplot(x = 'survival_status', y = 'operation_year', data = df)  #box-plot showing a relation between the year of operation and survival status
plt.subplot(1,3,2)
plt.subplot(1,3,3)
sns.boxplot(x = 'survival_status', y = 'positive_axillary_nodes', data = df)  #box-plot showing a relation between the positive auxillary nodes and survival status
plt.subplot(1,3,2)
plt.show()
```

```python
plt.figure(figsize = (15, 4))
plt.subplot(1,3,1)
sns.violinplot(x = 'survival_status', y = 'patient_age', data = df) #violin plot showing a relation between the patient's age and survival status
plt.subplot(1,3,2)
sns.violinplot(x = 'survival_status', y = 'operation_year', data = df) #violin plot showing a relation between the year of operation and survival status
plt.subplot(1,3,3)
sns.violinplot(x = 'survival_status', y = 'positive_axillary_nodes', data = df) #violin plot showing a relation between the positive auxillary nodes and survival status
plt.show()
```

```python
sns.set_style('darkgrid')
sns.pairplot(df, hue = 'survival_status')
plt.show()
```
